```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
flights = sns.load_dataset('flights')
flights.head()
```

```python
fc=flights.corr()
```

```python

sns.heatmap(fc,cmap=sns.color_palette("Set2"))
```

```python
pt=fc.pivot_table(values="passengers",index="month",columns="year")
sns.heatmap(pt)
```

```python
sns.clustermap(fc)
```

```python
tips=sns.load_dataset("tips")
tips.head()
```

```python
sns.lmplot(data=tips,x="tip",y="total_bill",hue="day",markers=["*",".","^",7])
```

```python
# using Grid column
sns.lmplot(data=tips,x="tip",y="total_bill",hue="day",markers=["*",".","^",7],col="sex")
```

```python
sns.lmplot(data=tips,x="tip",y="total_bill",hue="day",markers=["*",".","^",7],row="day",height=3)
```

```python
plt.figure(figsize=(3,3))
sns.countplot(data=tips,x="sex",)
sns.set_style("darkgrid")
sns.despine(left=True)
sns.set_context("talk",font_scale=0.1)
```
