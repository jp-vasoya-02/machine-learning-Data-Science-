```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = sns.load_dataset("penguins")
df.head()
```

```python
g=sns.PairGrid(df,palette="pastel",hue="sex")
g.map_diag(plt.hist,)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

```python
sns.pairplot(df,hue='species',palette='rainbow',kind="hist")
```

```python
fg=sns.FacetGrid(data=df,row="island",col="species")
fg.map(sns.histplot,"flipper_length_mm")
```

```python
jg=sns.JointGrid(x="species",y="sex",data=df)
jg.plot(sns.scatterplot,sns.histplot)
```
