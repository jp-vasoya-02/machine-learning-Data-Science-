```python
planets=sns.load_dataset("planets")
planets.head()
```

```python
plt.style.use("seaborn-v0_8-white")
planets.plot.area(alpha=0.5,cmap="coolwarm")
# plt.style.available
```

```python

planets.head(10).plot.bar(x="method",y="distance")
# plt.xlabel("")
plt.tight_layout()
plt.xticks([])
# 
# plt.axhline('off')
```

```python
planets["mass"].plot.hist()
```

```python
planets.index.name="index"
```

```python
planets.reset_index().plot.line(x="index",y="orbital_period",figsize=(10,2),lw=1)
plt.xlabel("-------------------->")
plt.ylabel("Orbital Period")
```
