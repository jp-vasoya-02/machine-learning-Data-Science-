```python
import pandas as pd
```

```python
df=pd.read_csv("Donation_data.csv").dropna(axis=1,thresh=4)
df
```

```python
df["Amount_number"]=float(df.loc[0,"Amount"].split("₹")[1].strip().replace(',',''))
```

```python
df["Amount_number"]=df["Amount"].apply(lambda a:float(a.split("₹")[1].strip().replace(',','')))
```
