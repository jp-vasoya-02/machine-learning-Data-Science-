```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
```

```python
fig=plt.figure()
axes=fig.add_axes([0.1,0.1,0.8,0.8])
x=np.arange(10)
y=x/2
axes.plot(x,y,"r")
axes.set_xlabel(' X Label') # Notice the use of set_ to begin methods
axes.set_ylabel('y Label')
axes.set_title('OOPS!')
plt.show()
```

```python
# INSIDE PLOT
fig=plt.figure()
axes=fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y,"y")
axes.set_xlabel("X axis")
axes.set_ylabel("Y axis")
axes2=fig.add_axes([0.2,0.6,0.3,0.3])
axes2.plot(x,y,"b")
axes2.set_xlabel("X axis")
axes2.set_ylabel("Y axis")
plt.show()
```

```python
# subplots

fig,axes=plt.subplots(nrows=2,ncols=1)
cno=1
for ax in axes:
    ax.plot(np.arange(20),np.random.randint(1,50,20),"g")
    ax.set_xlabel("X axis")
    ax.set_ylabel("Y axis")
    ax.set_title(f"Chart{cno}")
    cno+=1
plt.tight_layout()
_=0
clr=["r","y"]
fig,axes=plt.subplots(nrows=1,ncols=2,figsize=(10,3))
for ax in axes:
    ax.plot(np.arange(20),np.random.randint(1,50,20),clr[_])
    ax.set_xlabel("X axis")
    ax.set_xlabel("X axis")
    ax.set_title(f"Chart{cno}")
    cno+=1
    _+=1   
del _,clr,cno
plt.tight_layout()
plt.show()
fig.savefig("multi-chart.svg")
```

```python
x=np.random.randint(1,100,20)
y=np.arange(20)
```

```python
fig,axes=plt.subplots()
axes.plot(y,x,"r--",label="very constistent",lw=3,marker=".",markerfacecolor="blue",markeredgecolor="yellow",markersize=8,markeredgewidth=4)
axes.plot(y,x*5,label="most popular",color="#0ba608",alpha=0.5,lw=2,ls="solid",marker="s",markerfacecolor="blue",markeredgecolor="yellow",)
plt.legend(loc='center left', bbox_to_anchor=(1, 0.1))
plt.show()
```

```python
fig,axes=plt.subplots(nrows=1,ncols=3,figsize=(12,4))
axes[0].plot(y,x,y**2,x**3,"r")
axes[0].set_title("Default axes range")
# axes[0].plot(y,x)
axes[1].plot(y,x,y**2,x**3,"g")
axes[1].axis('tight')
axes[1].set_title("Tight axes range")
axes[2].plot(y,x,y**2,x**3,"y")
# axes[2].axis('tight')
axes[2].set_title("Custom axes range")
axes[2].set_ylim([20000,1000000])
axes[2].set_xlim([0,400])
plt.tight_layout()
plt.show()
```
